```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_boston
```

```python
boston=load_boston()
X=boston.data
y=boston.target
columns=boston.feature_names
```

```python
df=pd.DataFrame(X,columns=columns)
df.head()
```

```python
df.info()
```

```python
num_cols=list(df.select_dtypes(exclude=["object"]).columns)
num_cols
```

## hist plot for loop

```python
for col in num_cols:
    plt.figure(figsize=(10,5))
    plt.xlabel(col)
    plt.hist(x=df[col])
```

## box plot for loop 

```python
for col in num_cols:
    plt.figure(figsize=(10,5))
    sns.boxplot(x=df[col])
```

## outlier treatment

```python
sns.boxplot(x=df["CRIM"])
```

```python
df["CRIM"].describe()
```

## box plot formula 

```python
Q1=df["CRIM"].quantile(0.25)
Q3=df["CRIM"].quantile(0.75)
IQR=Q3-Q1
df=df[~((df["CRIM"]<(Q1-1.5*IQR))| (df["CRIM"]>(Q3+1.5*IQR)))]
```

```python
sns.boxplot(x=df["CRIM"])
```

```python
df["CRIM"].describe()
```
